```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap

data = pd.read_csv("traffic_result.csv")


features = data[['volume', 'occupancy', 'speed']]
labels = data['District']


districts = labels.unique()
cmap = ListedColormap([plt.cm.get_cmap('tab20')(i) for i in range(len(districts))])


pca = PCA(n_components=2)
X_pca = pca.fit_transform(features)


fig, ax = plt.subplots(figsize=(20,8))
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=labels.map({dist: i for i, dist in enumerate(districts)}), cmap=cmap)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title('Volume vs District')


handles = [plt.Line2D([], [], color=cmap(i), marker='o', linestyle='', label=dist) for i, dist in enumerate(districts)]
ax.legend(handles=handles, loc='upper right', title='Districts')

plt.show()
```
